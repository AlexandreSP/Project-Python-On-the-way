## 2. EDA Part 2 

What is the export value (US$) over the period?

```python
export_year= exportacao_full.groupby("Year").sum()
export_year = export_year [["US$ FOB"]]

#criar uma funcao de formatacao
def format(value):
    return f"$ {value:,.2f}"
#aplicar a formatacao 
export_year ["US$ FOB"]= export_year["US$ FOB"].map(format)
display(export_year)
```

In which month was there the highest export value (US$) over the period?

```python

export_Month= exportacao_full.groupby("Month").sum()
export_Month = export_Month [["US$ FOB"]]

export_Month = export_Month.sort_values('US$ FOB', ascending= False )

#criar uma funcao de formatacao
def format(value):
    return f"$ {value:,.2f}"
#aplicar a formatacao 
export_Month ["US$ FOB"]= export_Month["US$ FOB"].map(format)
display(export_Month)
```

What are the most exported products (US$) over the period?

```python
# Quais os produtos mais exportados (US$) ao longo do periodo?

export_product =exportacao_full.groupby("SH4 Description").sum()
export_product = export_product [["US$ FOB"]]

export_product = export_product.sort_values('US$ FOB', ascending= False )


export_product ["US$ FOB"]= export_product["US$ FOB"].map(format)



display(export_product)

```

What is the value in (US$) of exports between the economic blocks over the period?

```python
 

export_block =exportacao_full.groupby("Economic_Block").sum()
export_block = export_block [["US$ FOB"]]

export_block = export_block.sort_values('US$ FOB', ascending= False )


export_block ["US$ FOB"]= export_block["US$ FOB"].map(format)



display(export_block)
```

Which countries bought the most from Brazil in (US$) over the period?

```python


export_country = exportacao_full.groupby("Country").sum()
export_country = export_country[['US$ FOB']]
export_country = export_country.sort_values('US$ FOB', ascending= False)



export_country["US$ FOB"] = export_country["US$ FOB"].map(format)

display(export_country)
```

Which cities in Brazil exported the most in (US$) over the period

```python
export_city = exportacao_full.groupby("City").sum()
export_city = export_city[['US$ FOB']]
export_city = export_city.sort_values('US$ FOB', ascending= False)



export_city["US$ FOB"] = export_city["US$ FOB"].map(format)

display(export_city)
```
